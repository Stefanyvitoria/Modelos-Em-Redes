```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.externals import joblib
```

```python
graph = joblib.load(r'/content/drive/MyDrive/Modelos-Em-Redes/graph.pkl')
```

## Nós e Enlaces

```python
n, m = graph.number_of_nodes(), graph.number_of_edges()
print(f'Quantidade de Nós: {n}')
print(f'Quantidade de Enlaces: {m}')
```

## Componentes

```python
componentes = sorted(nx.connected_components(graph), key=len, reverse=True)
print(f'Quantidade de componentes conectadas: {len(componentes)}')
```

```python
len_componentes = []
for c in componentes:
  len_componentes.append(len(c))

print(len_componentes)

for i in range(1,9):
  print(f'componentes de tamanho {i}: {len_componentes.count(i)}')
```

Maior componente conectada

```python
mc = graph.subgraph(componentes[0])
n_mc, m_mc = mc.number_of_nodes(), mc.number_of_edges()
print(f'Tamanho da maior componente (em nós): {len(mc)} ')
print(f'Quantidade de Enlaces: {m_mc}')
```

## Densidade

Densidade da rede

```python
density = nx.density(graph)
print(f'Densidade: {density}')
```

Densidade da maior componente conectada

```python
density_mc = nx.density(mc)
print(f'Densidade: {density_mc}')
```

## Grau

### Sequencia dos graus

Sequencia dos graus da Rede

```python
deg = sorted([d for _,d in nx.degree(graph)],reverse=True)
print(f'Sequencia dos graus:\n{deg}')
```

Sequencia dos graus da Maior Componente

```python
deg_mc = sorted([d for _,d in nx.degree(mc)], reverse=True)
print(f'Sequencia dos graus:\n{deg_mc}')
```

### Grau Médio 

Grau Médio da Rede

```python
lambd = np.average(deg)
print(f'Grau médio: {lambd}')
```

Grau Médio da maior componete

```python
lambd_mc = np.average(deg_mc)
print(f'Grau médio: {lambd_mc}')
```

### Grau máximo (Hub)

Hub da Rede

```python
hubs = [] 
max = np.max(deg)
for _,d in nx.degree(graph):
  if d == max:
    hubs.append(d)
print(f'Hubs: {hubs}')
```

Hub da Maior componente

```python
hubs_mc = [] 
max_mc = np.max(deg_mc)
for _,d in nx.degree(mc):
  if d == max_mc:
    hubs_mc.append(d)
print(f'Hubs: {hubs_mc}')
```

###Porcentagem de nós abaixo/acima do grau médio da MC

```python
abaixo = acima = 0
for no in deg_mc:
  if no < lambd_mc: abaixo += 1
  elif no > lambd_mc: acima += 1

print(f'nós abaixo de lambda: {abaixo}')
print(f'nós acima de lambda: {acima}')
```

### Distribuição do grau

#### Frequência do grau

Frequência do grau da rede

```python
h = [ i/graph.number_of_nodes() for i in nx.degree_histogram(graph) ]
print(f'Frequência: {h}')
```

Frequência do grau da maior componente


```python
h_mc = [ i/mc.number_of_nodes() for i in nx.degree_histogram(mc) ]
print(f'Frequência: {h_mc}')
```

#### Entropia

Entropia da rede

```python
I = - np.nansum(h*np.log2(h))
print(f'Entropia = {str(I)}\n')
```

Entropia da maior componente


```python
I_mc = - np.nansum(h_mc*np.log2(h_mc))
print(f'Entropia = {str(I_mc)}\n')
```

#### Histograma

Histograma da rede

```python
plt.hist(deg, weights= np.ones_like(deg) / len(deg))
plt.xlabel('Grau')
plt.ylabel('Probabilidade')
plt.show()
```

Histograma da maior componente


```python
plt.hist(deg_mc, weights= np.ones_like(deg_mc) / len(deg_mc))
plt.xlabel('Grau')
plt.ylabel('Probabilidade')
plt.show()
```

## CC

### CC

CC da rede

```python
cc = nx.average_clustering(graph)
print(f'CC: {str(cc)}')
```

CC da maior componente

```python
cc_mc = nx.average_clustering(mc)
print(f'CC_mc: {str(cc_mc)}')
```

### nós com CC = 1

CC =1 da rede

```python
cc_1 = [nx.clustering(graph, no) for no in graph.nodes]

print(f'Quantidade de nós com Coeficiente de aglomeração igual a 1: {cc_1.count(1)}')
```

CC=1 da maior componente

```python
cc_1_mc = [nx.clustering(mc, no) for no in mc.nodes]

print(f'Quantidade de nós com Coeficiente de aglomeração igual a 1: {cc_1_mc.count(1)}')
```

## TMC

Como essa medida precisa de uma rede conectada e este projeto é para fins educacionais usei a maior componente.

```python
tmc_mc = nx.average_shortest_path_length(mc)
print(f'TMC: {tmc_mc}')
```

## Eficiência dos enlaces

```python
E_mc = 1 - tmc_mc/m_mc
print(f"Eficiência dos enlaces: {str(E_mc)}")
```
